# Navigation

Jupyter notebooks work by entering and processing cells. Cells are typed, and there are a few basic types. The most popular are code and markdown. We will enter and execture our Python code in Code cells and use Markdown cells to enter documentation using the same Markdown language we've been using on GitHub. 

The third type of cell is NBConvert - these are cells that are not interpreted/processed by Jupyter. 

This cell is a markdown cell. 

To change cell type, go to Cell->Cell Type Menu

NOTE: Hitting escape in a cell will turn the call border blue. When in this mode, you can enter commands. For instance, you can change a cell to Markdown by simply entering M. 
    
    For a complete list of these types of commands see  https://www.cheatography.com/weidadeyue/cheat-sheets/jupyter-notebook/

### Lists

#### Raw Markdown of Lists:
```

2. Another item
  * Unordered sub-list. 
1. Actual numbers don't matter, just that it's a number
  1. Ordered sub-list
4. And another item.
```
#### Rendered Markdown of Lists:

1. First ordered list item
2. Another item
   * Unordered sub-list. 
1. Actual numbers don't matter, just that it's a number
   1. Ordered sub-list
4. And another item.

## Plotting with Bokeh

### Scatter Plots

````python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook

output_notebook()

p = figure(plot_width=400, plot_height=400)

# add a circle renderer with a size, color, and alpha
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="navy", alpha=0.5)

# show the results
show(p)
````

````python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook

output_notebook()

p = figure(plot_width=400, plot_height=400)

# add a square renderer with a size, color, and alpha
p.square([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="olive", alpha=0.5)

# show the results
show(p)
````

### Line Plots

````python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
output_notebook()

# prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a new plot with a title and axis labels
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x, y, legend="Temp.", line_width=2)

# show the results
t=show(p)
````

````python
from bokeh.plotting import figure, output_file, show

output_notebook()

p = figure(plot_width=400, plot_height=400)

p.multi_line([[1, 3, 2], [3, 4, 6, 6]], [[2, 1, 4], [4, 7, 8, 5]],
             color=["firebrick", "navy"], alpha=[0.8, 0.3], line_width=4)

show(p)
````
